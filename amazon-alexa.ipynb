## we will make a pipeline to remove (punctuation , stop words) and then tokenize it using (Count vectorizer) 

```python
# now we have to make data preprocessing

# clean data .. punct

import string
string.punctuation
```

```python
# then remove stopwords
import nltk 
nltk.download('stopwords')

from nltk.corpus import stopwords
# stopwords.words('english')
```

```python
# then tokenize the text using count tokenizer

import sklearn 
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
```

```python
# Create a pipeline to preform task in a sequence

def message_cleaned(message):
    punc_removed = [ char for char in message if char not in string.punctuation]
    punc_removed_and_combined = ''.join(punc_removed)
    stop_word_removed = [word for word in punc_removed_and_combined.split() if word.lower() not in stopwords.words('english')]
    return stop_word_removed
```

```python
reviews_cleaned_df = df['verified_reviews'].apply(message_cleaned) 
```

```python
print(reviews_cleaned_df[5])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Define the cleaning pipeline we defined earlier
vectorizer = CountVectorizer(analyzer = message_cleaned, dtype = np.uint8) #very important step 
reviews_countvectorizer = vectorizer.fit_transform(df['verified_reviews'])
```

```python
X = pd.DataFrame(reviews_countvectorizer.toarray())
```

```python
X[:4]

X['']
```

```python
X['rating'] = df['rating'] 
```

```python
X[:3]
```

```python
X.drop('feedback' , axis = 1 , inplace = True)
```

```python
df.head()
```

```python
y = df['feedback']
```

```python
X.drop('rating' , axis = 1 , inplace = True)
```

```python
X.shape
```

```python
y.shape
```

```python
# time to get the model 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

from sklearn.naive_bayes import MultinomialNB
NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
y_predict_test = NB_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict_test))
```

```python
# the weighted average in 91% among all --> not bad 
```

```python
# try gradient boosting
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot = True)

print(classification_report(y_test, y_pred))
```
